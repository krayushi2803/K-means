```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv("iris.csv")
y1=df["petal.length"]
y2=df["petal.width"]
```

```python
data=list(zip(y1,y2))
model=KMeans(n_clusters=2)
model.fit(data)
plt.scatter(y1,y2,c=model.labels_)
plt.show()
```
